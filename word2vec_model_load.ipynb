```python
# include the libraries

import pandas as pd
import csv
from collections import Counter
from gensim.models import word2vec

from word2vec_functions import *
```

```python
## Load the trained model

model = word2vec.Word2Vec.load('yelp_word2vec')
```

```python
## Load Lucky's market reviews

lucky_reviews = []
with open('lucky_market_problem.txt', 'rb') as f:
    reader = csv.reader(f)
    lucky_reviews = [row[0] for row in reader]
```

```python
# reviews to list of word list
lucky_sentences = []  # Initialize an empty list of sentences

# print "Parsing sentences from training set"
for review in lucky_reviews:
    lucky_sentences += review_to_sentences(review, tokenizer, True)
```

```python
## Most common words

words_to_count = [word for sentence in lucky_sentences for word in sentence ]
counter_lucky = Counter(words_to_count)
```

```python
print '1 The total number of reviews: %d\n' % len(lucky_reviews)

print '2 Most common words among the reviews:'
counter_lucky.most_common(10)
```

```python
## Gather all reviews that contains the key_word and all similar words from the trained model

print '3 Find all the similar words to "cashier" through the trained model:'
model.most_similar('cashier')
```

```python
print '4 Gather all the reviews for topic "cashier":'
key_word_lookup(lucky_reviews, model, key_word='cashier')
```
